```python
# Заменяем буквы на числа

for col_name in data.columns:
    # Do something with the column data
    encoder = preprocessing.LabelEncoder() # строки в числа
    data[col_name] = encoder.fit_transform(data[col_name])
    print(f"Column Name: {col_name}")
    print(data[col_name].value_counts())
    print("-----------------")
```
